# Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
future=data.drop('PRICE',axis=1)
target=data['PRICE']
```

```python
x_train,x_test,y_train,y_test=train_test_split(pd.get_dummies(future),target,test_size=0.2)
```

```python
x_train
```

```python
y_test
```

# Fitting XG BOOST REGRESSOR to the training set

```python
import xgboost as xgb
```

```python
xg=xgb.XGBRegressor()
```

```python
model=xg.fit(x_train,y_train)
```

# Predicting the test set results

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

# Machine performance score

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
model.score(x_test,y_pred)
```

# R-squared

```python
r2_score(y_test,y_pred)
```

```python
model.score(x_test,y_test)
```
