# Text Preprocessing

# Removing HTML Tags

```python
import re
```

```python
def clean_html(x):
    clean=re.compile('<.*?>')
    return re.sub(clean,'',x)
```

```python
data['review']=data['review'].apply(clean_html)
```

```python
data['review'].iloc[2]
```

# Converting Every Word to Lowercase 

```python
def convert_to_lower(x):
    return x.lower()
```

```python
data['review']=data['review'].apply(convert_to_lower)
```

```python
data['review'].iloc[2]
```

```python
data['review'].iloc[11]
```

# Removing Special Characters

```python
def remove_special(text):
    x=''
    
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
data['review']=data['review'].apply(remove_special)
```

# Removing Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
a=stopwords.words('english')
```

```python
a
```

```python
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in a:
            x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
data['review']=data['review'].apply(remove_stopwords)
```

```python
data['review'].iloc[11]
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem_words(text):
    y=[]
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
data['review']=data['review'].apply(stem_words)
```

```python
data['review'].iloc[11]
```

```python
def join_back(list_input):
    return ' '.join(list_input)
```

```python
data['review']=data['review'].apply(join_back)
```

# Model Output

```python
def predict_sentiment(text):
    text_clean=clean_html(text)
    text_clean=convert_to_lower(text_clean)
    text_clean=remove_special(text_clean)
    text_clean=remove_stopwords(text_clean)
    text_clean=stem_words(text_clean)
    text_clean=join_back(text_clean)
    text_vec=cv.transform([text_clean])
    prediction=nb2.predict(text_vec)
    return "Positive" if prediction[0]==1 else "Negative"
```

```python
input_text="This movie made me so happy.<bt /> A must-see!"
print(predict_sentiment(input_text))
```

```python
input_text="I hated this movie.**The acting was awful.//"
print(predict_sentiment(input_text))
```
