# Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
future=data_new.drop('MarketValue',axis=1)
target=data_new['MarketValue']
```

```python
X_train,X_test,y_train,y_test=train_test_split(future,target,test_size=0.2)
```

```python
X_train
```

```python
y_train
```

# Fitting Ada Boost Regressor to the training set

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
Ad=AdaBoostRegressor()
```

# Predicting the test set results

```python
model=Ad.fit(X_train,y_train)
y_pred=model.predict(X_test)
    
```

```python
y_pred
```

```python
len(y_pred)
```

# Machine performance score

```python
model.score(X_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
